```python
from sklearn import datasets
import pandas as pd
import numpy as np
```

```python
iris_datasets = datasets.load_iris(as_frame=True)
iris_datasets
```

```python
iris_df = iris_datasets['frame']
iris_df = iris_df[iris_df['target'].isin((0,1))]
```

```python
def log_loss(y_pred: pd.Series, y: pd.Series) -> float:
    """
    Log Loss
    """

    return -np.mean(y * np.log(y_pred) + (1 - y) * np.log(1 - y_pred))
```

```python
def L_Reg(X: pd.Series, y: pd.Series, lr: float, steps: int) -> pd.Series:
    """
    Linear Regression
    """
    a = 0
    b = 0
    for i in range(steps):
        y_pred = 1  / (1 + np.e ** -(a * X + b))
        da = np.sum((y_pred - y) * X) / len(X)
        db = np.sum((y_pred - y)) / len(X)
        
        a -= lr * da
        b -= lr * db
        
        print(f"Step {i:4}: a = {a:.5f}, b = {b:.5f}, log_loss = {log_loss(y_pred, y):.8f}")
    
    
```

```python
L_Reg(iris_df['petal width (cm)'], iris_df['target'], 0.1, 10000)
```
